```python
import cv2
from PIL import Image
import cv2
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("patrick.png")
b, g, r = cv2.split(img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calculate the histogram of each channel
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])
hist_gray = cv2.calcHist([gray], [0], None, [256], [0, 256])

# # Plot the histograms
# plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(hist_b, color='blue')
plt.title("Blue Histogram")
plt.xlim([0, 256])

plt.subplot(2, 2, 2)
plt.plot(hist_g, color='green')
plt.title("Green Histogram")
plt.xlim([0, 256])

plt.subplot(2, 2, 3)
plt.plot(hist_r, color='red')
plt.title("Red Histogram")
plt.xlim([0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_gray, color='black')
plt.title("Gray Histogram")
plt.xlim([0, 256])

plt.tight_layout()
plt.show()

# Average intensity
average_intensity = np.mean(gray)
print("Average Intensity:", average_intensity)

# Image contrast
def calculateContrast(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    std = np.std(img)
    return std

contrast = calculateContrast(img)
print("Contrast:", contrast)

# PENJELASAN
# a.) histogram RGB
# Histogram RGB adalah representasi statistik dari distribusi intensitas warna merah (R), hijau (G), dan biru (B) dalam sebuah gambar berwarna.
# Ini mengukur seberapa sering intensitas piksel tertentu muncul dalam masing-masing saluran warna (R, G, B). Histogram RGB berguna untuk 
# menganalisis distribusi warna dalam gambar dan dapat digunakan untuk berbagai tujuan, seperti peningkatan kontras atau koreksi warna.

# b.) histogram greyscale
# Histogram grayscale adalah representasi statistik dari distribusi intensitas piksel dalam gambar grayscale.Ini mengukur seberapa sering 
# intensitas piksel tertentu muncul dalam skala abu-abu (dari 0 hingga 255 dalam gambar 8-bit). Histogram grayscale berguna untuk 
# menganalisis tingkat kecerahan atau kontras dalam gambar dan dapat digunakan dalam pemrosesan gambar seperti segmentasi objek atau 
# peningkatan citra.

# c.) Average Intensity
# Intensitas rata-rata adalah nilai rata-rata dari intensitas piksel dalam gambar atau saluran warna tertentu.Untuk gambar grayscale, 
# ini adalah rata-rata dari semua nilai piksel dalam gambar. Ini mengukur tingkat kecerahan umum gambar. Untuk gambar berwarna, Dapat 
# dihitung rata-rata intensitas untuk masing-masing saluran warna (R, G, B) secara terpisah. ntensitas rata-rata berguna untuk mengukur 
# tingkat kecerahan keseluruhan suatu gambar.

# d.) Image Contrast
# Kontras gambar mengukur perbedaan antara intensitas piksel yang berdekatan dalam gambar. Semakin besar perbedaan intensitas antar piksel, 
# semakin tinggi kontras gambarnya. Kontras gambar digunakan untuk mengukur sejauh mana detail dan perbedaan antar objek dalam gambar dapat
# dilihat dengan jelas. Ini dapat dihitung dengan berbagai metode, salah satunya adalah menghitung deviasi standar (standar deviasi) dari 
# intensitas piksel dalam gambar. Semakin tinggi deviasi standarnya, semakin tinggi kontrasnya.
```
