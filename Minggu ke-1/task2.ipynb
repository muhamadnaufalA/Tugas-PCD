```python
from PIL import Image
import cv2
from IPython.display import display
import numpy as np
import copy
```

```python
img = cv2.imread('R.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display (Image.fromarray(img))
```

```python
lower_red = np.array([255,0,0])
upper_red = np.array([255,0,0])

mask = cv2.inRange(img, lower_red, upper_red)

keliling = np.sum(mask==255)
sisi = keliling/4
luas = sisi*sisi

logo = Image.open('R.png')
print("Keliling pixel segiempat merah:", keliling)
print("Luas pixel segiempat merah:", luas)
print("Filename:", logo.filename)
print("Format:", logo.format)
print("Mode:", logo.mode)
print("Size:", logo.size)
print("Widht:", logo.width)
print("Height:", logo.height)
```

```python
img[mask > 0] = [255,255,255]

border_color = (255, 0 , 0)

border_width = 10

borderer_image = cv2.copyMakeBorder(img,border_width, border_width, border_width,
                                    border_width, cv2.BORDER_CONSTANT, value=border_color)

display (Image.fromarray(borderer_image))
```

```python
lower_red = np.array([0,0,254])
upper_red = np.array([0,0,255])

mask = cv2.inRange(img, lower_red, upper_red)

badge_image_without_red = cv2.bitwise_and(img,img, mask =~ mask)

bingkai = np.zeros_like(img)
tinggi, lebar, _ = img.shape
ketebalan_bingkai = 10

cv2.rectangle (bingkai, (0,0), (lebar,tinggi), (0,0,255), ketebalan_bingkai)

gambar_akhir = cv2.add(badge_image_without_red, bingkai)

cv2.imshow('HASIL', gambar_akhir)
cv2.waitKey(0)
cv2.destroyAllWindows
```
