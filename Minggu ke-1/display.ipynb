```python
from PIL import Image
import cv2
from IPython.display import display

img = cv2.imread('kadal.jpeg')
display(Image.fromarray(img))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display (Image.fromarray(img))
```

```python
r , g, b = img [0, 0]
print ("RED:", r)
print("GREEN:", g)
print("BLUE:", b)
```

```python
r , g, b = cv2.split(img)
print ("RED:", r)
print("GREEN:", g)
print("BLUE:", b)
```

```python
pil_016 = Image.fromarray(r)
display(pil_016)
```

```python
pil_016 = Image.fromarray(g)
display(pil_016)
```

```python
pil_016 = Image.fromarray(b)
display(pil_016)
```

```python
merged = cv2.merge ([b,g,r])
pil_016 = Image.fromarray (merged)
display (pil_016)
```

```python
merged = cv2.merge ([r,g,b])
pil_016 = Image.fromarray (merged)
display (pil_016)
```
